```python
import matplotlib.pyplot as plt


x = [32, 16, 8, 4]
y = [
    17.073698043823242,
    17.09531021118164,
    24.78927993774414,
    33.801597595214844
]

plt.figure()
plt.plot(x, y, marker='o')
plt.xlabel("Number of Channels")
plt.ylabel("CER")
plt.title("The Relationship Between the Number of Channels and CER")
plt.grid(True)

plt.show()
```

```python
import matplotlib.pyplot as plt


x = [16, 12, 8, 4]
y = [
    17.073698043823242,
    19.667171478271484,
    21.979684829711914,
    76.1400451660156
]

plt.figure()
plt.plot(x, y, marker='o')
plt.xlabel("Amount of Training Data (session)")
plt.ylabel("CER")
plt.title("The Relationship Between the Amount of Training Data and CER")
plt.grid(True)
plt.show()
```
